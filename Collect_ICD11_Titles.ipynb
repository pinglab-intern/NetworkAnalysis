```python
import json as json
```

```python
with open("Data/MMS_rv.json", 'r') as f:
    Data =  json.load(f)
```

```python
title = []
def extract_cardio(cardio):
    for doc in Data:
        if(doc['ID'] == cardio):
            title.append(doc['title'])
            l = len(doc['childs'])
            if(l > 0):
                for i in range(l):
                    extract_cardio(doc['childs'][i])

for doc in Data:
    if(doc['ID'] == "426429380"):
        l = len(doc['childs'])
        if(l > 0):
            title.append(doc['title'])
            for i in range(l):
                extract_cardio(doc['childs'][i])
```

```python
with open('cvd.txt', 'w') as f:
    for item in title:
        f.write(item)
        f.write('\n')
```
